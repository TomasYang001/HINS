```python
import pandas as pd
import numpy as np
```

```python
Lung1 = pd.read_csv("Lung1.csv")
Lung1 = Lung1.rename(columns={'Patient ID':'Patient', 'Amino Acid (protein)': 'Protein Change', 'Consequence':'Class'})
Lung1 = Lung1[["Patient", "Gene", "Protein Change", "Class"]]

Lung2 = pd.read_csv("Lung2.csv")
Lung2 = Lung2.rename(columns={"Hugo_Symbol":"Gene", "pair_id":"Patient", "Protein_Change":"Protein Change", "Variant_Classification":"Class"})
Lung2 = Lung2[["Patient", "Gene", "Protein Change", "Class"]]

Mel1 = pd.read_csv("Mel1.csv")
Mel1 = Mel1.rename(columns={"Sample":"Patient", "Variant_Class":"Class", "aaannotation":"Protein Change", "Hugo_Symbol":"Gene"})
Mel1 = Mel1[["Patient", "Gene", "Protein Change", "Class"]]

Mel2 = pd.read_csv("Mel2.csv")
Mel2 = Mel2.rename(columns={"Sample":"Patient", "MutType":"Class", "Aamut":"Protein Change"})
Mel2 = Mel2[["Patient", "Gene", "Protein Change", "Class"]]

Gene_list = ["TP53", "KRAS", "BRAF", "NRAS", "EGFR", "BCLAF1", "PIK3CA",\
             "ATM", "HLA-A", "HLA-B", "B2M", "JAK1", " NF1", "CDKN2A",\
             "TAP1", "TERT", "CTNNB1", "PPP6C", "STK11", "FAT1", "IFNGR1", "RAC1"]

df_combined = pd.concat([Lung1, Lung2 ,Mel1, Mel2], axis=0, ignore_index=True)
Gene_data = pd.DataFrame()
for i in range(len(Gene_list)):
    temp = df_combined[df_combined["Gene"] == Gene_list[i]]
    Gene_data = pd.concat([Gene_data, temp], axis = 0, ignore_index=True)
```

```python
# ID = Gene_data["Patient"]
# ID_list = list(ID)
# unique_ID = []
# [unique_ID.append(item) for item in ID_list if item not in unique_ID]
# print(len(unique_ID))

Gene_descending = Gene_data.sort_values(by='Patient', ascending=False)
Pats = list(Gene_descending["Patient"])
frequency = pd.Series(Pats).value_counts()
Patient_list = frequency[frequency >= 3] #取出现3种以上突变的患者
Patient_list = list(Patient_list.index)
Patient_list

Mutation_data = pd.DataFrame()
for Pt in range(len(Patient_list)):
    pt_temp = Gene_data[Gene_data["Patient"] == Patient_list[Pt]]
    Mutation_data = pd.concat([Mutation_data, pt_temp], axis = 0, ignore_index=True)

matrix = Mutation_data.pivot_table(index='Gene', columns='Patient', values='Class', aggfunc='sum')
```

```python
Clinical = pd.read_csv("/Users/tomasyang/Documents/YANG/科研/雪鸢组/杨强/HINS/R/Cohorts/Results.csv")
Clinical = Clinical[["Sample.ID", "Cancer", "Smoking", "Response"]]

Clinical_list = [s+"-T1" if s.startswith("SU2CLC") else s for s in Clinical["Sample.ID"]] #与Mutation数据中的Patient ID保持一致

Clinical["Sample.ID"] = Clinical_list
Clinical_data = pd.DataFrame()
for pt in range(len(Patient_list)):
    Cl_temp = Clinical[Clinical["Sample.ID"] == Patient_list[pt]]
    Clinical_data = pd.concat([Clinical_data, Cl_temp], axis = 0, ignore_index=True)

```

```python
Gene_data.to_csv("Gene_data.CSV", index = None) #基因总体数据
matrix.to_csv("matrix.CSV")#所求基因突变数据
Clinical_data.to_csv("Clinical data.csv", index = None)#对应临床数据
```
